```python
#dependancies
import pandas as pd

# File to Load 
file_to_load = "../Data/Video_Games_Sales_Dec_2016.csv"

# Read file and store in a DF
game_data_df = pd.read_csv(file_to_load)
```

```python
#check to see data has loaded
game_data_df.head()
```

```python
#drop all rows with missing data. Many of the older games are missing information like ratings
dupe_drop_df = game_data_df.dropna()
dupe_drop_df.head()
```

```python
#dropping columns that are not needed. It will not help analysis to know how many votes there are in this instance
dropped_df = dupe_drop_df.drop(columns=['Critic_Count', 'User_Count'])
dropped_df.head()
```

```python
#user score was an object, converting to a useful format so I can multipy the values by 10 in line with the critic score
dropped_df['User_Score'] = dropped_df['User_Score'].astype(float)
dropped_df.head()
```

```python
dropped_df['User_Score'] = dropped_df['User_Score'] * 10
dropped_df.head()
```

```python
clean_videogame_data_df = dropped_df
```

```python
#confirming that all numeric values are of the correct data types
clean_videogame_data_df.dtypes
```

```python
# determining the name of the file
file_name = '../Data/clean_videogame_data.xlsx'
  
# saving the excel for excel analysis
clean_videogame_data_df.to_excel(file_name, index=False)
print('DataFrame is written to Excel File successfully.')
```

```python
#saving as CSV for python analysis
file_name_1 = '../Data/clean_videogame_data_csv.csv'
clean_videogame_data_df.to_csv(file_name_1, index=False)
```
