```python
#dependancies
import pandas as pd

# File to Load 
file_to_load = "../Data/clean_videogame_data_csv.csv"

# Read file and store in a DF
game_data_df = pd.read_csv(file_to_load)
game_data_df.head()
```

```python
#check the number of unique titles
#this drastically reduced number is explained by games being released on multiple platforms and some titles are exclusive.
clean_count = game_data_df['Name'].nunique()
clean_count
```

```python
#checking the number of platforms, this number was surprisingly high until I realised that there were several handheld consoles in the mix
platform_count = game_data_df['Platform'].nunique()
platform_count
```

```python
console_group = game_data_df.groupby(['Platform'])
pd.DataFrame(console_group.size().reset_index(name = "Console_Count"))
```
